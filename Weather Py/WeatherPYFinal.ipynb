```python
import pandas as pd
import numpy as np
import matplotlib as plt
from ratelimit import limits
import scipy.stats as sts
import json
import requests
from citipy import citipy
api_key = "2d7bc754b1f208baf1d97b9f731cceb6"
```

```python
#ranges
larange = (-25, 50)
lonrange = (-70, 125)

base_url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
query_url = f"{base_url}appid={api_key}&units={units}&q="
```

```python
latlon = []
cities = []

lat = np.random.uniform(low =- 90.000, high = 90.000, size = 1250)
lon = np.random.uniform(low =- 180.000, high = 180.000, size = 1250)
latlon = zip(lat,lon)

for latlon in latlon:
    city = citipy.nearest_city(latlon[0], latlon[1]).city_name
    
    if city not in cities:
        cities.append(city)
len(cities)
```

```python
print(cities)
```

```python
#build lists for data
citame = []
latlist = []
lonlist = []
mte = []
huity = []
clness = []
wied = []
coury = []
date = []

cone = 0
sene = 1

for index, city in enumerate(cities, start = 1):
    try:
        response = requests.get(query_url + city).json()
        citame.append(response['name'])
        latlist.append(response['coord']['lat'])
        lonlist.append(response['coord']['lon'])
        mte.append(response['main']['temp_max'])
        huity.append(response['main']['humidity'])
        clness.append(response['weather']['main'])
        wied.append(response['wind']['speed'])
        coury.append(response['sys']['country'])
        date.append(response['dt'])
        if cone > 10:
            cone = 0
            sene = sene + 1
        else:
            cone = cone + 1
        print(f'Searching for {cone} of {sene} | {city}')
    except Exception:
        print("City not available or API is stupid")
print(" Data Search Complete")
```

```python
clc = pd.DataFrame({"City": citame, "Lat": latlist, "Lng": lonlist, 
                           "Max Temp": mte, "Humidity": huity, "Cloudiness": clness, 
                           "Wind Speed": wied, "Country": coury, "Date": date})
clc.head()
```
